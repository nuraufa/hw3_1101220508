```python
import pandas as pd
```

```python
vf = pd.read_csv("LB_TA.txt", delimiter = r"\s+")
vf
```

menampilkan data dalam bentuk txt

```python
vf = vf.drop_duplicates()
vf
```

menghapus baris yang memiliki nilai NaN pada kolom at

```python
vf = vf[vf['at'].notna()]
vf
```

menggunakan .agg() untuk menghitung rata-rata 'at' berdasarkan 'tic' dan reset index

```python
vf = vf.groupby('tic').agg({'at': 'mean'}).reset_index()
vf
```

mengubah nama kolom at menjadi avg_at

```python
vf = vf.rename(columns={'at': 'avg_at'})
vf
```
