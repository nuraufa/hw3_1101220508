```python
import pandas as pd
```

```python
kl = pd.read_excel(r"List Customer Call.xlsx")
kl
```

```python
kl = kl.drop_duplicates()
kl
```

```python
kl = kl.drop(columns = "Not_Useful_Column")
kl
```

```python
kl["Last_Name"] = kl["Last_Name"].str.strip("abc")
kl
```

menghapus kolom pada Last Name

```python
import re
# Mengganti NaN dengan string kosong
kl["Phone_Number"] = kl["Phone_Number"].fillna('')
# Mengganti 'nan--', 'Na--', dan 'n/a' dengan string kosong menggunakan regex
kl["Phone_Number"] = kl["Phone_Number"].replace(r'nan--|Na--|n/a', '', regex=True)
# Menggunakan regex untuk menghapus semua karakter non-angka
kl["Phone_Number"] = kl["Phone_Number"].replace(r'[^0-9]', '', regex=True)
kl
```

menggunakan fungsi apply dan lambda untuk membagi string berdasarkan koma

```python
kl[['Street_Address', 'State', 'Zip_Code']] = kl['Address'].apply(lambda x: pd.Series(x.split(',', 2)))
kl
```

menggunakan fungsi .map() untuk mengganti nilai 'Yes' dan 'No' menjadi 'Y' dan 'N'

```python
kl["Do_Not_Contact"] = kl["Do_Not_Contact"].map({'Yes': 'Y', 'No': 'N'}).fillna(kl["Do_Not_Contact"])
kl["Paying Customer"] = kl["Paying Customer"].map({'Yes': 'Y', 'No': 'N'}).fillna(kl["Paying Customer"])
kl
```

```python
kl = kl.fillna('')
kl
```

menggunakan .loc untuk mengganti nilai 'N/a' dengan string kosong

```python
kl = kl.replace('N/a'," ")
kl
```

menghapus baris yang memiliki nilai 'Y' pada kolom 'Do_Not_Contact' menggunakan kondisi boolean

```python
kl = kl[kl["Do_Not_Contact"] != 'Y']
kl
```
